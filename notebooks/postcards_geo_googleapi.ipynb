```python
import requests
import pandas as pd

# set up API parameters
api_key = 'AIzaSyAeiqC8X9lyI4xTE_pgVcxsoV7d1_g-b18'
base_url = 'https://maps.googleapis.com/maps/api/geocode/json'

# create a function to geocode a single city
def geocode(city):
    params = {
        'address': city,
        'key': api_key
    }
    response = requests.get(base_url, params=params)
    data = response.json()
    if data['status'] == 'OK':
        lat = data['results'][0]['geometry']['location']['lat']
        lng = data['results'][0]['geometry']['location']['lng']
        return lat, lng
    else:
        return None, None
```
