```python
img_names = []
with open('img_ids.txt') as in_file:
    for line in in_file:
        line = line.strip()
        img_names.append(line)
```

```python
import pandas as pd
```

```python
df1 = pd.read_csv('../data/processed/metadata.csv')
```

```python
df1['id'] = df1.index
```

```python
img_names_= ['_'.join(n.split('/')[-1].split('_')[1:]).split('.')[0] for n in img_names]
```

```python
df = df1.copy()
```

```python
df['id'] = -1
```

```python
df['NAME_'] = df['NAME'].apply(lambda x: x.split('.')[0])
```

```python
for i, n in enumerate(img_names_):
    df.loc[df['NAME_'] == n, 'id'] = i
```
