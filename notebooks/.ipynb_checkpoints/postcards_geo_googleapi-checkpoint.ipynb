```python
import requests
import pandas as pd

# set up API parameters
api_key = 'AIzaSyAeiqC8X9lyI4xTE_pgVcxsoV7d1_g-b18'
base_url = 'https://maps.googleapis.com/maps/api/geocode/json'

# create a function to geocode a single city
def geocode(city):
    params = {
        'address': city,
        'key': api_key
    }
    response = requests.get(base_url, params=params)
    data = response.json()
    if data['status'] == 'OK':
        lat = data['results'][0]['geometry']['location']['lat']
        lng = data['results'][0]['geometry']['location']['lng']
        return lat, lng
    else:
        return None, None
```

```python
import time
import json
import requests
```

```python
def get_openstreetmap_data(citylist):

    payload = {'q': ' '.join(citylist), 'format': 'json'}
    r = requests.get('https://nominatim.openstreetmap.org/search', params=payload)
    openstreetmap_data = json.loads(r.text)
    time.sleep(1)

    return openstreetmap_data
```

```python
def get_openstreetmap_data1(city):
    payload = {'q': city, 'format': 'json'}
    r = requests.get('https://nominatim.openstreetmap.org/search', params=payload)
    time.sleep(1)
    return json.loads(r.text)[0]
```
